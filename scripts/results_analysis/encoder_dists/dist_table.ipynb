```python
import json
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'font.size': 22})
```

```python
exp_folder = "../../../logs/analysis/version_9"
langs = json.load(open(os.path.join(exp_folder, "langs.json")))
branches = [0, 16]
```

```python
langs
```

```python
for branch in branches:
    dists = json.load(open(os.path.join(exp_folder, f"tsne/b{branch}", "euclidean_dists.json")))
    print(f"Branch {branch}")
    print(f"| | {' | '.join([f'{lang}' for lang in langs])} |")
    print(f"| | {' | '.join([':---:' for lang in langs])} |")
    for i, row in enumerate(dists):
        print(f"| {langs[i]} | {' | '.join([f'{val:.1f}' for val in row])} |")
```

```python
for branch in branches:
    dists = json.load(open(os.path.join(exp_folder, f"tsne/b{branch}", "euclidean_dists.json")))
    plt.figure(figsize=(10, 10))
    plt.imshow(dists)
    plt.title(f"Euclidean distances on branch {branch}")
    plt.gca().xaxis.tick_top()
    plt.xticks(range(len(langs)), langs, rotation=45)
    plt.yticks(range(len(langs)), langs)
    plt.colorbar()
    plt.clim(0, 150)
```

```python
for branch in branches:
    dists = json.load(open(os.path.join(exp_folder, f"tsne/b{branch}", "euclidean_dists.json")))
    print(branch, np.array(dists).mean())
```
