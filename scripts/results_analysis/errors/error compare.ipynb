```python
import json
import os
```

```python
exp_dir_1 = "../../../logs/en/analysis/version_21"
exp_dir_2 = "../../../logs/en/analysis/version_22"

error_types = ["replace", "insert", "delete"]

errors_1 = json.load(open(os.path.join(exp_dir_1, "errors.json")))
errors_2 = json.load(open(os.path.join(exp_dir_2, "errors.json")))
```

```python
def occurrences(errors, error_type):
    """Return the words, number of errors and occurrences of the given error type."""
    words = []
    for word, data in errors.items():
        if error_type in data:
            if data["occurrences"] > 5:
                words.append((word, data[error_type], data["occurrences"]))
    words.sort(key=lambda x: x[1], reverse=True)
    return words
```

```python
for word in occurrences(errors_1["test_sc.txt"], "replace"):
    print(word)
```

```python
word_1  = occurrences(errors_1["test_sc.txt"], "replace")
word_2  = occurrences(errors_2["test_sc.txt"], "replace")
comp = []
difference = 0
for word_f in word_1:
    for word_x in word_2:
       if word_f[0] == word_x[0]:
           diff = abs(word_f[1] - word_x[1])
           if diff > 0:
            comp.append((word_f[0], diff, word_f[2]))
            difference += diff 
            break
    
    comp.sort(key=lambda x: x[1], reverse=True)

difference
```
