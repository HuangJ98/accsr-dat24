```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
sns.set_theme(context='notebook')
```

```python
colours = ["navy","green","darkred","cornflowerblue","limegreen","red"]
counter = 0
folder_ac = "../Results/FCCN/Accented"

plt.figure(figsize=(10, 7))
for file in sorted(os.listdir(folder_ac)):
    df = pd.read_csv(os.path.join(folder_ac,file))
    label = file.split(".")[1]
    plt.plot(df["Step"],df["Value"], color = colours[counter], label=label) 
    counter += 1
  

plt.xlabel("Epoch") 
plt.ylabel("Score") 
#plt.title("FCCN scoring for accented") 
  
# Adding legend, which helps us recognize the curve according to it's color 
plt.legend(loc = "lower right") 
  
# To load the display window 
plt.show() 
```
